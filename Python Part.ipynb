Q1.	Write a function that returns the following sum: 1+2+3+...+n

```python
def f1(n):
    sum=0
    for i in range(1,n+1):
        sum += i
    return sum
```

```python
#Test function f1
f1(5)    #Output should be 15
```

Q2. A function that deduces the factorial of any number nâˆˆN, only using a while loop.

```python
def f2(n):
    fac=1
    count=1
    while count<=n:
        fac = fac*count
        count += 1
    return fac
```

```python
#Test function f2
f2(5)     #Output should be 120
```

```python
f2(0)
```

Q3. A program that uses a loop to find the minimum of a vector x, without using any predefined functions like min()/sort().

```python
def f3(x):
    m=x[0]
    i = 1
    while i< (len(x)):
        if x[i]< m:
            m = x[i]
        i += 1
    return m
```

```python
#Test function f3
x= [5,3,7,1,9,2]
f3(x)    #Output should be 1
```

```python
f3([3,2,5,1])
```

```python
#Q4.
import matplotlib.pyplot as plt

data = {'Lack of exercise': 0.38,'Easy access to junk food': 0.23,
        'Genetics': 0.12,'Eating unhealthy food': 0.09,
        'Medical condition': 0.08,'Overeating': 0.07}

factors =list(data.keys())
percentages = list(data.values())

# Plotting a bar chart
plt.bar(factors, percentages)

# Adding labels and title
plt.xlabel('Factors')
plt.ylabel('Percentage')
plt.title('The most important contributing factors to child\'s weight problem')
plt.xticks(rotation=45) #The x-axis labels are long and may overlap so rotating them can make them more readable.

# Display the plot
plt.show()
```

It would be reasonable that:
"Lack of exercise" with "Overeating" might be grouped into a single category named 'Unhealthy Lifestyle Habits' or 'Behavioral Factors'. Both factors contribute to an imbalance between calorie intake and expenditure, leading to weight gain

and "Easy access to junk food" and "Eating unhealthy food" can be aggregated into the category 'Poor Dietary Choices' or 'Unhealthy Eating Habits'. Both factors involve consuming high-calorie, low-nutrient foods, which can contribute to weight gain.

While we keep 'Genetics' and 'Medical condition' separate.

OR another way to combine some of those factors into a single category:
'Lack of Exercise', 'Easy access to junk food', 'Eating unhealthy food' and 'Overeating might be combined into single category named 'Lifestyle choice ' because they all relate to lifestyle choices.



```python
#Q5. Using matplotlib to construct a histogram 

import matplotlib.pyplot as plt

wind_speed = [30.3, 39.0, 33.9, 38.6, 44.6, 31.4, 26.7, 51.9, 31.9,
              27.2, 52.9, 45.8, 63.3, 36.0, 64.0, 31.4, 42.2, 41.1,
              37.0, 34.4, 35.5, 62.2, 30.3, 40.0, 36.0, 39.4, 34.4,
              28.3, 39.1, 55.0, 35.0, 28.8, 25.7, 62.7, 32.4, 31.9,
              37.5, 31.5, 32.0, 35.5, 37.5, 41.0, 37.5, 48.6, 28.1]

# Plotting a basic histogram
plt.hist(wind_speed, bins=8, edgecolor='black')

# Adding labels and title
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Frequency')
plt.title('The annual maximum wind speed in Hong Kong in 45 years')

# Display the plot
plt.show()
```

```python
#Q5. Using seaborn to construct a histogram and add a density plot

import seaborn as sns
import matplotlib.pyplot as plt

# Creating a customized histogram with a density plot
sns.histplot(wind_speed,bins=8, kde=True)

# Adding labels and title
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Frequency')
plt.title('The annual maximum wind speed in Hong Kong in 45 years')

# Display the plot
plt.show()
```

From the above histograms, we can see that:

The histogram appears to be positively skewed, as the tail extends to the right. - The histogram is unimodal since there is only one peak. 

In conclusion, the histogram for the annual maximum wind speed data is positively skewed and unimodal.
